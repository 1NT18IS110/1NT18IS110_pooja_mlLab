# K-Means clustering

```python
import math


def eucledian(x,y,ncx,ncy):
    return math.sqrt((ncx-x)**2 +(ncy-y)**2)

def kMeans():
    cluster=list()
    ocx=[0,0,0]
    ocy=[0,0,0]
    
    ncx=[6.2,6.6,6.5]
    ncy=[3.2,3.7,3.0]
    
    
    points = [[5.9,3.2],
           [4.6,2.9],
           [6.2,2.8],
           [4.7,3.2],
           [5.5,4.2],
           [5.0,3.0],
           [4.9,3.1],
           [6.7,3.1],
           [5.1,3.8],
           [6.0,3.0]]
    
    diffx=1
    diffy=1
    itr=1
    while(diffx !=0 and diffy!=0):
        print("In iteration ",itr)
        ocx=ncx
        ocy=ncy
        cluster=[]
        
        for i in range(len(points)):
            dist=[]
            for j in range(3):
                dist.append(eucledian(points[i][0],points[i][1],ocx[j],ocy[j]))
            m=dist.index(min(dist))
            
            cluster.append(m+1)
            
        print("cluster points:",cluster)
        
        print("Old centroid x and y:")
        print(ocx)
        print(ocy)
        
        ncx=[]
        ncy=[]
        
        c1x,c1y,c2x,c2y,c3x,c3y=0,0,0,0,0,0
        
        for i in range(len(cluster)):
            if cluster[i]==1:
                c1x+=points[i][0]
                c1y+=points[i][1]
            if cluster[i]==2:
                c2x+=points[i][0]
                c2y+=points[i][1]
            if cluster[i]==3:
                c3x+=points[i][0]
                c3y+=points[i][1]
        if(cluster.count(1)!=0):
            ncx.append(c1x/cluster.count(1))
            ncy.append(c1y/cluster.count(1))
        if(cluster.count(2)!=0):
            ncx.append(c2x/cluster.count(2))
            ncy.append(c2y/cluster.count(2))
            
        if(cluster.count(3)!=0):
            ncx.append(c3x/cluster.count(3))
            ncy.append(c3y/cluster.count(3))
        
        diffx=0
        diffy=0
        
        print("New centroid x and y:")
        print(ncx)
        print(ncy)
        
        for i in range(0,3):
            diffx+=abs((ncx[i]-ocx[i]))
            diffy+=abs((ncy[i]-ocy[i]))  
        itr+=1


kMeans()
```
